## Выравнивание гистограммы:

```python
def histeg(im,nbr_bins=256):
    imhist,bins = histogram(im.flatten(),nbr_bins)
    cdf = imhist.cumsum()
    cdf = 255 * cdf / cdf[-1]

    im2 = interp(im.flatten(),bins[:-1],cdf)
    return im2.reshape(im.shape), cdf
    
```

## Усреднение изображений:

```python
def compute_average(imlist):
    averageim = array(Image.open(imlist[0]), 'f')
    for imname in imlist[1:]:
        try:
            averageim += array(Image.open(imname))
        except:
            print(imname,'...пропущено')
            averageim /= len(imlist)
    
    return array(averageim, 'uint8')
```

## Метод главных компонент для изображений

```python
from math import sqrt
from PIL import Image
from numpy import *

def pea(X):

    num_data, dim = X.shape

    mean_X = X.mean(axis=0)
    X = X - mean_X
    if dim>num_data:

        M = dot(X,X.T)
        e,EV = linalg.eigh(M)
        tmp = dot(X.T,EV).T
        V = tmp[::-1]

        S = sqrt(e)[::-1]

        for i in range(V.shape[1]):
            V[:,i] /= S

    else:
        U,S,V = linalg.svd(X)
        V = V[:num_data]
    
    return V,S,mean_X
```

## Очистка изображения от шумов:

```python
import numpy as np

def denoise(im, U_init, tolerance=0.1, tau=0.125, tv_weight=100):
    """Реализация модели очистки от шумов Рудина-Ошера-Фатеми (ROF) с использованием численного алгоритма, приведенного в формуле А. Шамболя d 2005.
Вход: зашумленное изображение (полутоновое), начальная гипотеза о U, вес члена, регуляризирующего ТУ, величина шага,допуск в условии остановки.
Выход: очищенное от шума и текстуры изображение, остаточная текстура. """

    m, n = im.shape
    U = U_init
    Px = im
    Py = im
    error = 1

    while error > tolerance:
        Uold = U
        GradUx = np.roll(U, -1, axis=1) - U
        GradUy = np.roll(U, -1, axis=0) - U

        PxNew = Px + (tau/tv_weight) * GradUx
        PyNew = Py + (tau/tv_weight) * GradUy
        NormNew = np.maximum(1, np.sqrt(PxNew**2 + PyNew**2))

        Px = PxNew / NormNew
        Py = PyNew / NormNew

        RxPx = np.roll(Px, 1, axis=1)
        RyPy = np.roll(Py, 1, axis=0)
        DivP = (Px - RxPx) + (Py - RyPy)
        U = im + tv_weight * DivP

        error = np.linalg.norm(U - Uold) / np.sqrt(n * m)

    return U, im - U
```

## 2. Реализуйте операцию нерезкого маскирования, для чего сначала размойте изображение, а затем вычтите размытое изображение из исходного. Это повышает резкость изображения. Попробуйте применить операцию к полутоновому и цветному изображению.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def unsharp_masking(image, sigma=1.0, strength=1.5):
    # Применяем гауссово размытие
    blurred = cv2.GaussianBlur(image, (0, 0), sigmaX=sigma, sigmaY=sigma)

    # Вычитаем размытое изображение из оригинала
    sharpened = cv2.addWeighted(image, 1.0 + strength, blurred, -strength, 0)

    return sharpened

# Загрузка полутонового изображения
gray_image = cv2.imread('img1_grey.jpg', cv2.IMREAD_GRAYSCALE)

# Загрузка цветного изображения
color_image = cv2.imread('img1.jpg')

# Параметры для операции нерезкого маскирования
sigma = 2.5
strength = 5.5

# Применяем операцию к полутоновому изображению
sharpened_gray = unsharp_masking(gray_image, sigma=sigma, strength=strength)

# Применяем операцию к цветному изображению
sharpened_color = unsharp_masking(color_image, sigma=sigma, strength=strength)

# Выводим результаты
plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(gray_image, cmap='gray')
plt.title('Исходное серое изображение')
plt.axis('off')

plt.subplot(222)
plt.imshow(sharpened_gray, cmap='gray')
plt.title('Резкое серое изображение')
plt.axis('off')

plt.subplot(223)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('Исходное цветное изображение')
plt.axis('off')

plt.subplot(224)
plt.imshow(cv2.cvtColor(sharpened_color, cv2.COLOR_BGR2RGB))
plt.title('Резкое цветное изображение')
plt.axis('off')

plt.show()
```

## 3. Альтернативой выравниванию гистограммы является метод нормировки изображения с помощью вычисления изображения-частного. Оно получается путем деления изображения на его размытый вариант . Реализуйте эту операцию и протестируйте на каких-нибудь примерах.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_division(image, sigma=1.0):
    # Размываем изображение
    blurred = cv2.GaussianBlur(image, (0, 0), sigmaX=sigma, sigmaY=sigma)
    
    # Производим операцию деления
    result = np.float32(image) / (np.float32(blurred) + 1.0)
    
    # Масштабируем результаты в диапазон [0, 255] и преобразуем в формат uint8
    result = np.clip(result * 255, 0, 255).astype(np.uint8)
    
    return result

# Загрузка изображения
image = cv2.imread('img1.jpg', cv2.IMREAD_GRAYSCALE)

# Параметры для операции нормировки изображения-частного
sigma = 150.0

# Применяем операцию нормировки изображения-частного
normalized_image = image_division(image, sigma=sigma)

# Выводим результаты
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(normalized_image, cmap='gray')
plt.title('Нормализованное изображение')
plt.axis('off')

plt.show()
```

## 4. Напишите функцию, которая находит контуры простых объектов внутри изображения (например, квадрата на белом фоне) с помощью градиентов.

```python
import cv2
import numpy as np

def find_simple_contours(image):
    # Преобразование изображения в оттенки серого
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Применение адаптивной пороговой обработки
    binary = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    
    # Поиск контуров на двоичном изображении
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Фильтрация контуров по форме
    filtered_contours = []
    for contour in contours:
        perimeter = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, 0.02 * perimeter, True)
        if len(approx) == 4:  # Если контур представляет собой прямоугольник
            filtered_contours.append(contour)
    
    # Отрисовка контуров и текста на исходном изображении
    result_image = image.copy()
    for i, contour in enumerate(filtered_contours):

        # Находим ограничивающий прямоугольник для контура
        x, y, w, h = cv2.boundingRect(contour)
        
        # Находим центр ограничивающего прямоугольника
        center_x = x + w // 2
        center_y = y + h // 2
        
        # Отрисовываем контур
        cv2.drawContours(result_image, [contour], -1, (0, 255, 0), 2)
        
        # Вычисляем размер текста
        text_size = cv2.getTextSize(f"Object {i+1}", cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)[0]
        
        # Вычисляем позицию текста так, чтобы его центр совпадал с центром контура
        text_x = center_x - text_size[0] // 2
        text_y = center_y + text_size[1] // 2  # Смещаем на высоту текста, чтобы центр текста совпал с центром контура
        
        # Добавляем текст в центре контура
        cv2.putText(result_image, f"Object {i+1}", (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
    
    return result_image, filtered_contours



# Пример использования функции на изображении с квадратом на белом фоне
image_with_square = cv2.imread('shapes.png')

# Находим контуры на изображении с квадратом
contoured_image, contours = find_simple_contours(image_with_square)

# Отображаем результаты
cv2.imshow('Contoured Image', contoured_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5. Воспользуйтесь модулем и направлением градиента для распознавания отрезков прямых внутри изображения. Оцените длину отрезков и параметры соответствующих прямых. Нанесите отрезки поверх изображения на графике.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_lines(image):
    # Преобразование изображения в оттенки серого
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Нахождение градиентов с помощью оператора Собеля
    gradient_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    gradient_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    
    # Нахождение модуля и направления градиента
    gradient_magnitude = np.sqrt(gradient_x ** 2 + gradient_y ** 2)
    gradient_direction = np.arctan2(gradient_y, gradient_x)
    
    # Применение пороговой обработки для выделения границ
    _, edges = cv2.threshold(gradient_magnitude, 50, 255, cv2.THRESH_BINARY)
    
    # Преобразование кантов в изображение формата uint8
    edges = np.uint8(edges)
    
    # Нахождение отрезков прямых с помощью преобразования Хафа
    lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=100, minLineLength=50, maxLineGap=5)
    
    # Нанесение отрезков на исходное изображение
    result_image = image.copy()
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(result_image, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    return result_image

# Загрузка изображения
image = cv2.imread('shapes.png')

# Распознавание отрезков прямых и их отображение
result_image = detect_lines(image)

# Отображение результатов
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

## 6. Примените функцию label() к бинаризованному изображению по своему выбору. Воспользовавшись гистограммами и результирующим меточным изображением, постройте график распределения размеров объектов внутри изображения.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Загрузка бинаризованного изображения
binary_image = cv2.imread('shapes.png', cv2.IMREAD_GRAYSCALE)

# Применение функции label() для маркировки объектов
labeled_image, num_features = ndimage.label(binary_image)

# Вычисление размеров объектов
sizes = ndimage.sum(binary_image, labeled_image, range(num_features + 1))

# Построение гистограммы распределения размеров объектов
plt.figure(figsize=(10, 5))
plt.hist(sizes[1:], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Распределение размеров объектов')
plt.xlabel('Размер')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

## 7. Поэкспериментируйте с последовательным применением морфологических операций к бинаризованному изображению по своему выбору. Подобрав параметры, при которых получается хороший результат, попробуйте применить функцию center_of_mass из модуля morphology для нахождения координат центра каждого объекта и нанесите найденные точки на изображение.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage.measure import label, regionprops

# Загрузка бинаризованного изображения
binary_image = cv2.imread('quad.jpg', cv2.IMREAD_GRAYSCALE)

# Применение расширения
kernel = np.ones((5, 5), np.uint8)
dilated_image = cv2.dilate(binary_image, kernel, iterations=1)

# Применение эрозии
eroded_image = cv2.erode(dilated_image, kernel, iterations=1)

# Маркировка объектов
labeled_image, num_features = label(eroded_image, connectivity=1, background=0, return_num=True)

# Нахождение координат центра
centers = []
for region in regionprops(labeled_image):
    center = region.centroid
    centers.append((int(center[1]), int(center[0])))  # Поменяли местами координаты для OpenCV (x, y)

# Отрисовка
result_image = cv2.cvtColor(binary_image, cv2.COLOR_GRAY2BGR)
for center in centers:
    cv2.circle(result_image, center, 10, (0, 0, 255), -1)  # Нарисуем красными кругами

# Отображение результатов
plt.figure(figsize=(10, 5))
plt.imshow(result_image)
plt.title('Центры объектов')
plt.axis('off')
plt.show()
```
